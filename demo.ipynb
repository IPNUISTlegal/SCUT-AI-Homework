```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from haze_removal import dehaze
```

## Read the origin image.

```python
img = io.imread('img/ny1.png')
plt.figure('origin')
plt.imshow(img)
```

## Remove haze. The image shown is the dark channel.

```python
I = np.asarray(img, dtype=np.float64)
result_raw, result_refined = dehaze(I)
```

## Estimate transmission.

```python
plt.figure('result')
plt.imshow(result_raw)
```

## Refine the transmission. Notice that the white edges disappear.

```python
plt.figure('refined')
plt.imshow(result_refined)
```
